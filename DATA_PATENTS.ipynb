```python
from bs4 import BeautifulSoup
import urllib3
import pandas as pd

#Dataframe object to store the scrapped details of all patents
data=pd.DataFrame(columns=['Patent ID','Title','Abstract','Inventors','Assignee','Filed'])


http = urllib3.PoolManager()


#List of Patent no's
List=['7976828','7867948','7811611','8202545','8202906','8012515','8802163', '8513456']

Count =0
for i in List:
    Count+=1
    
    #USPTO DATABASE
    url='http://patft.uspto.gov/netacgi/nph-Parser?Sect1=PTO1&Sect2=HITOFF&d=PALL&p=1&u=%2Fnetahtml%2FPTO%2Fsrchnum.htm&r=1&f=G&l=50&s1='+i+'.PN.&OS=PN/'+i+'&RS=PN/'+i
    
    print("Scrapping url "+str(Count)+"\n"+url+'\n')
    response = http.request('GET', url)
    soup = BeautifulSoup(response.data)
    
    
    #series object to collect Title,Abstract,Inventors,Assigne
    data_row={}
    data_row['Patent ID']=i
    
    #script to collect title
    title=soup.find('font',size='+1')
    data_row['Title']=title.text.replace('\n','')
    
    #script to collect Abstract
    Abstract=soup.p.text
    data_row['Abstract']=Abstract.replace('\n    ','')

    #script to collect Assigne,Filling Date and Inventors
    table=soup('table')[3]
    children = table.findChildren("tr" , recursive=False)
    for child in children:
        if("Assignee:" in child.th.text):
            data_row['Assignee']=child.td.text.strip('\n\r\t": ')
        if("Filed" in child.th.text):
            data_row['Filed']=child.td.text.strip('\n\r\t": ')
        if("Inventors" in child.th.text):
            data_row['Inventors']=child.td.text.strip('\n\r\t": ')
    
    
    #If assigne is not given on database set to none
    if('Assignee' not in data_row.keys()):
        data_row['Assignee']='None'

        
        
        
    data=data.append(data_row,ignore_index=True)    
    
print('DATA SCRAPPED')
```
